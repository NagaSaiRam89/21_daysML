## **How has the distribution of content ratings changed over time?**

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', dayfirst=True)
```

```python
df['year_added'] = df['date_added'].dt.year
display(df.head())
```

```python
ratings_over_time = df.groupby(['year_added', 'rating']).size().reset_index(name='count')
display(ratings_over_time.head())
```

```python
plt.figure(figsize=(14, 7))
sns.histplot(data=ratings_over_time, x='year_added', weights='count', hue='rating', multiple='stack', shrink=0.8)
plt.title('Distribution of Content Ratings Over Time')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.show()
```

**Insigts** :
From the above insights we can say that the rating of shows and movies increase over time from 2015

## **What are the most common word pairs or phrases in content descriptions?**

```python
import string
from nltk.corpus import stopwords

# Download stopwords if not already downloaded
import nltk
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

def clean_text(text):
    """Converts text to lowercase, removes punctuation and stop words."""
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

df['cleaned_description'] = df['description'].apply(clean_text)
display(df[['description', 'cleaned_description']].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(2, 2))
bigrams_matrix = vectorizer.fit_transform(df['cleaned_description'])
bigram_counts = pd.DataFrame(bigrams_matrix.sum(axis=0), columns=vectorizer.get_feature_names_out()).T
bigram_counts = bigram_counts.rename(columns={0: 'count'})
top_bigrams = bigram_counts.sort_values(by='count', ascending=False).head(10)
display(top_bigrams)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_bigrams.index, y='count', data=top_bigrams, palette='viridis')
plt.title('Top 10 Most Frequent Bigrams in Netflix Content Descriptions')
plt.xlabel('Bigram')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Insigths:The most frequent bigrams provide insights into common themes and subjects present in the Netflix content descriptions, such as education ("high school", "college student") and relationships ("young woman", "young man").

# **Who are the top directors on Netflix?**

```python
director_counts = df['director'].value_counts().reset_index()
director_counts.columns = ['director', 'count']
display(director_counts.head())
```

```python
n = 10
top_directors = director_counts.head(n)
top_directors.drop(top_directors[top_directors['director'] == 'Unknown'].index, inplace=True)
display(top_directors)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='director', y='count', data=top_directors, palette='viridis')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```
